```python
import pandas as pd
import numpy as np
import os
from tqdm import tqdm
from googletrans import Translator
from time import sleep 
import random
from random import shuffle
import re
import string
```

# 実験

```python
languages = ['es', 'fr', 'de', 'ja']
df = pd.read_csv("../data/train.csv").loc[:19, :]
trans_target = list(df.index)
random.shuffle(trans_target)
aug_num = 80
random.seed(42)
translator = Translator()

idx = 0
for i in range(aug_num + 1):
    if idx < len(df):
        translated = translator.translate(df.loc[trans_target[idx], "description"], dest=languages[i // len(df)]).text
        re_translated = translator.translate(translated, dest="en").text
        print(translated)
        print(re_translated)
        print(languages[i // len(df)])
        idx += 1
    else:
        random.shuffle(trans_target)
        idx = 0
```
